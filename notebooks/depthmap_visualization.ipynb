# Depth Map 시각화
1. depth map의 사이즈와 RGB 사이즈 일치시키기.
    - misalignment로 오류 발생시 depth registration 방식을 접근할 것. 
    - 뎁스맵을 RGB 해상도에 맞게 업샘플링(보간법) 사용
    - Camera Intrinsics를 기반으로 3D 포인트로 복원
    - 포인트에 RGB 색상 맵핑

```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
figure = plt.figure(figsize=(15,5))
plotnum=1
for num in range(1,6):
    json_path = f"YOUR DEPTH MAP PATH/{num}.json"

    with open(json_path, "r") as f:
        data = json.load(f)

    width = data["Width"]
    height = data["Height"]
    depth = np.array(data['Depth']).reshape(height, width)
    depth_rot = np.rot90(depth, -1)

    figure.add_subplot(1,5,plotnum)
    plt.imshow(depth_rot, cmap='gist_gray_r')  
    # plt.colorbar()  
    # plt.title(f'{num+1}')
    plt.axis("off")
    figure.show()
    plotnum+=1
```
